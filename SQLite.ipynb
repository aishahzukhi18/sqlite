## LIKE Operator

```python
#find any values that start with 'a'

%sql SELECT * FROM Customers WHERE ContactName LIKE 'a%' 
```

```python
#find any values that end with 'a'

%sql SELECT * FROM Customers WHERE ContactName LIKE '%a'
```

```python
#find any value that start with a and at leat 3 characters in length

%sql SELECT * FROM Customers WHERE ContactName LIKE 'a__%'
```

```python
#find any value that does not start with 'a'

%sql SELECT * FROM Customers WHERE NOT ContactName LIKE 'a%'
```

## BETWEEN Operator

```python
%%sql
SELECT * FROM Products
WHERE UnitPrice BETWEEN 5.0 AND 15.0
ORDER BY Unitprice DESC
```

```python
%%sql
SELECT * FROM Orders
WHERE OrderDate BETWEEN '1996-07-16' AND '1996-08-02';
```

## INNER JOIN

```python
%%sql
SELECT o.OrderID, o.OrderDate, c.CompanyName
FROM Customers AS c
INNER JOIN Orders as o 
ON c.CustomerID=o.CustomerID
WHERE c.CompanyName="Around the Horn"
```

## LEFT Join

```python
%%sql
SELECT Customers.CompanyName, Orders.OrderID
FROM Customers
LEFT JOIN Orders ON Customers.CustomerID=Orders.CustomerID
```

```python
%%sql
SELECT Customers.CompanyName, Orders.OrderID
FROM Customers
FULL OUTER JOIN Orders ON Customers.CustomerID=Orders.CustomerID
ORDER BY Customers.CompanyName;
```

## HAVING Clause

```python
%%sql
SELECT Country, count(*) as [Number of Customers] FROM Customers
GROUP BY Country
HAVING [Number of Customers] > 5
ORDER BY [Number of Customers] DESC
```
